### 3) FEATURE SELECTION

#### Correlation search for numerical attributes with target variable

```python
x = train.select_dtypes(include=np.number)
x = x.drop("SalePrice", axis=1)
y = train.loc[:,'SalePrice']

attributes = []
cor_values = []
p_values = []

for i in range (x.shape[1]):
    attributes.append(x.columns[i])
    cor_val, p_val = stats.pearsonr(x.iloc[:,i],y)
    cor_values.append(cor_val)
    p_values.append(p_val)

df1 = pd.DataFrame(attributes, columns = ["Attributes"])
df2 = pd.DataFrame(cor_values, columns = ["Pearson Corr."])
df3 = pd.DataFrame(p_values, columns = ["P-Values"])

df2 = df2.abs()

df  = pd.concat([df1,df2,df3], axis=1)
df  = df.sort_values(["Pearson Corr."], ascending = False)
df  = df.reset_index(drop=True)

df
```

```python
#Drop the attributes which have p-values larger than 5%

train = train.drop(['LowQualFinSF','MiscVal','YrSold','BsmtHalfBath','Id','BsmtFinSF2','MoSold','3SsnPorch'], axis=1)
test = test.drop(['LowQualFinSF','MiscVal','YrSold','BsmtHalfBath','Id','BsmtFinSF2','MoSold','3SsnPorch'], axis=1)
```

#### Correlation search for categorical attributes with target variable

```python
x = train.select_dtypes(include=object)
y = train.loc[:,'SalePrice']
attributes = []
cor_values = []
p_values = []

for i in range (15):
    attributes.append(x.columns[i])
    cor_val, p_val = stats.kendalltau(x.iloc[:,i],y)
    cor_values.append(cor_val)
    p_values.append(p_val)

df1 = pd.DataFrame(attributes, columns = ["Attributes"])
df2 = pd.DataFrame(cor_values, columns = ["Kendalltau Corr."])
df3 = pd.DataFrame(p_values, columns = ["P-Values"])

df2 = df2.abs()

df  = pd.concat([df1,df2,df3], axis=1)
df  = df.sort_values(["Kendalltau Corr."], ascending = False)
df  = df.reset_index(drop=True)

df
```

```python
#Drop the attributes which have p-values larger than 5%

train = train.drop(['Utilities','LandContour','Street'], axis=1)
test = test.drop(['Utilities','LandContour','Street'], axis=1)
```

### 4) Decision tree regression algorithm modelling
